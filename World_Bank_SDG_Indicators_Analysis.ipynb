## 1. Comparing continent/region/income/ for selected indicator

```python
# list used: indicator_list

# Display a menu of elements
print("Select an element:")
for index, element in enumerate(indicator_list, start=1):
    print(f"{index}. {element}")

# Get user input for selection
try:
    selection = int(input("Select the indicator index for which you want the comparison for the regions: "))

    # Check if the selection is within valid range
    if 1 <= selection <= len(indicator_list):
        selected_element = indicator_list[selection - 1]
        print("\n")
        print(f"\033[1;31mYou selected:\033[0m \033[1;34m{selected_element}\033[0m")
    else:
        print("Invalid selection. Please choose a valid number.")
except ValueError:
    print("Invalid input. Please enter a number.")


# working on the dataframe based on the user input
sel_list = ['Continent or region wise data','Income wise data']

# Display a menu of elements
print("Select an indicator:")
for index2, element2 in enumerate(sel_list, start=1):
    print(f"{index2}. {element2}")

try:
    selection2 = int(input("Please select which data you want to work on and type the number"))

    # Check if the selection is within valid range
    if 1 <= selection2 <= len(sel_list):
        selected_element2 = sel_list[selection2 - 1]
        print("\n")
        print(f"\033[1;31mYou selected:\033[0m \033[1;34m{selected_element2}\033[0m")
    else:
        print("Invalid selection. Please choose a valid number.")
except ValueError:
    print("Invalid input. Please enter a number.")

if selected_element2 == 'Continent or region wise data':
    df_for_user_indicator = continent_df[continent_df['Indicator Name']==selected_element]
elif selected_element2== 'Income wise data':
    df_for_user_indicator = income_df[income_df['Indicator Name']==selected_element]
else:
    print("Data don't exist")

df_for_user_indicator = df_for_user_indicator.drop(['Indicator Name'], axis=1)

df_for_user_indicator_T = df_for_user_indicator.T

df_for_user_indicator_T.columns = df_for_user_indicator_T.iloc[0]

df_for_user_indicator_T = df_for_user_indicator_T.iloc[1:]

df_for_user_indicator_T = df_for_user_indicator_T.reset_index().rename(columns={'index':'Year'}).rename_axis('index')

# Remove the name from the index
df_for_user_indicator_T.index.name = None

# To remove the name from the columns as well
df_for_user_indicator_T.columns.name = None

df_for_user_indicator_T = df_for_user_indicator_T.astype('int')
df_for_user_indicator_T['Year'] = df_for_user_indicator_T['Year'].astype('str')

column_list = list(df_for_user_indicator_T.columns)
del column_list[0]


from IPython.display import display

display(df_for_user_indicator_T)


## Graph kind 1--------------------------------------------------------------------------------##
import plotly.express as px

# Column to plot against all other columns
column_to_plot = 'Year'

# Extract all columns except the one to be plotted
other_columns = [col for col in column_list if col != column_to_plot]

# Create an interactive line plot with Plotly
fig = px.line(df_for_user_indicator_T, x='Year', y=other_columns, title=f'{selected_element} for regions of the world vs year', labels={'Year': 'Year', 'value': selected_element})
fig.update_xaxes(tickangle=45)
fig.show()
##--------------------------------------------------------------------------------##


##  Graph kind 2--------------------------------------------------------------------------------##

import plotly.express as px

# Column to plot against all other columns
column_to_plot = 'Year'

# Extract all columns except the one to be plotted
other_columns = [col for col in column_list if col != column_to_plot]

# Create a long-form DataFrame for Plotly
df_long = df_for_user_indicator_T.melt(id_vars=['Year'], value_vars=other_columns, var_name='Country/Region', value_name=selected_element)

# Create an interactive scatter plot with Plotly
fig = px.scatter(df_long, x='Year', y=selected_element, color='Country/Region', title=f'{selected_element} for regions of the world vs year')
fig.update_xaxes(tickangle=45)
fig.show()
##--------------------------------------------------------------------------------##

## Graph kind 3--------------------------------------------------------------------------------##

import seaborn as sns
# Column to plot against all other columns
column_to_plot = 'Year'

# Extract all columns except the one to be plotted
other_columns = [col for col in column_list if col != column_to_plot]

plt.figure(figsize=(12,25))  # Create a single figure for all lines

for elem in other_columns:
    sns.lineplot(x='Year', y=elem, data=df_for_user_indicator_T)

# Add labels to the end of the lines
for elem in other_columns:
    x, y = df_for_user_indicator_T['Year'].iloc[-1], df_for_user_indicator_T[elem].iloc[-1]
    plt.text(x, y, elem, ha='left', va='center')


plt.xticks(rotation=90)
plt.xlabel('Year')
plt.ylabel(f'{selected_element}')
plt.title(f'{selected_element} for {selected_element2} wise countries vs year')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
##--------------------------------------------------------------------------------##
```

# 2. Comparing indicator change for selected continent/regions/income/country.

```python
len(country_list)
```

```python
# working on the dataframe based on the user input

# Display a menu of elements
print("Select a country:")
for index3, element3 in enumerate(country_list, start=1):
    print(f"{index3}. {element3}")

try:
    selection3 = int(input("Please select country index for which you want the data"))

    # Check if the selection is within valid range
    if 1 <= selection3 <= len(country_list):
        selected_element3 = country_list[selection3 - 1]
        print("\n")
        print(f"\033[1;31mYou selected:\033[0m \033[1;34m{selected_element3}\033[0m")
    else:
        print("Invalid selection. Please choose a valid number.")
except ValueError:
    print("Invalid input. Please enter a number.")


locals()[selected_element3] = country_df[country_df['Country Name']==selected_element3]

locals()[selected_element3+'_T'] = locals()[selected_element3].T

locals()[selected_element3+'_T'].columns = locals()[selected_element3+'_T'].iloc[1]

locals()[selected_element3+'_T'] = locals()[selected_element3+'_T'].iloc[2:]

locals()[selected_element3+'_T'] = locals()[selected_element3+'_T'].reset_index().rename(columns={'index':'Year'}).rename_axis('index')



# Remove the name from the index
locals()[selected_element3+'_T'].index.name = None

# To remove the name from the columns as well
locals()[selected_element3+'_T'].columns.name = None

locals()[selected_element3+'_T'] = locals()[selected_element3+'_T'].astype('int')
locals()[selected_element3+'_T']['Year'] = locals()[selected_element3+'_T']['Year'].astype('str')


# Display a menu of elements
print("Select any 5 indicators to display the growth for the selected country:")
for index4, element4 in enumerate(indicator_list, start=1):
    print(f"{index4}. {element4}")

try:
    ind = [int(input(f'Please enter the indicator {i} index from list')) for i in range(5)]

    selected_ind_list = [indicator_list[ind[n]-1] for n in list(range(5))]
    print("\n")
    print(f"\033[1;31mYou selected:\033[0m \033[1;34m{selected_ind_list}\033[0m")

except ValueError:
    print("Invalid input. Please enter a number.")

from IPython.display import display
final_df = locals()[selected_element3+'_T'][['Year'] + selected_ind_list]
display(final_df)

## Graph kind 1--------------------------------------------------------------------------------##
import plotly.express as px

# Column to plot against all other columns
column_to_plot = 'Year'

# Extract all columns except the one to be plotted
other_columns = [col for col in selected_ind_list]

# Create an interactive line plot with Plotly
fig = px.line(final_df, x='Year', y=other_columns, title=f'The growth of 5 selected indicators for the country: {selected_element3}', labels={'Year': 'Year', 'value': '% of total(refer Legend)'})
fig.update_xaxes(tickangle=45)
fig.show()
##--------------------------------------------------------------------------------##

##  Graph kind 2--------------------------------------------------------------------------------##

import plotly.express as px

# Column to plot against all other columns
column_to_plot = 'Year'

# Extract all columns except the one to be plotted
other_columns = [col for col in selected_ind_list if col != column_to_plot]

# Create a long-form DataFrame for Plotly
df_long = final_df.melt(id_vars=['Year'], value_vars=other_columns, var_name='Indicator Legend', value_name='% of total(refer Indicator Legend)')

#Create an interactive scatter plot with Plotly

fig = px.scatter(df_long, x='Year', y='% of total(refer Indicator Legend)', color='Indicator Legend', title=f'The growth of 5 selected indicators for the country: {selected_element3}')
fig.update_layout(width=1200,height=400)
fig.update_xaxes(tickangle=30)
fig.show()
#--------------------------------------------------------------------------------##

## Graph kind 3--------------------------------------------------------------------------------##

import seaborn as sns
# Column to plot against all other columns
column_to_plot = 'Year'

# Extract all columns except the one to be plotted
other_columns = [col for col in selected_ind_list if col != column_to_plot]

plt.figure(figsize=(15,10))  # Create a single figure for all lines

for elem in other_columns:
    sns.lineplot(x='Year', y=elem, data=final_df)

# Add labels to the end of the lines
for elem in other_columns:
    x, y = final_df['Year'].iloc[-1], final_df[elem].iloc[-1]
    plt.text(x, y, elem, ha='left', va='center')


plt.xticks(rotation=90)
plt.xlabel('Year')
plt.ylabel(f'% of total(refer Indicator Legend)')
plt.title(f'The growth of 5 selected indicators for the country: {selected_element3}')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
##--------------------------------------------------------------------------------##
```
